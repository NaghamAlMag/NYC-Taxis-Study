We proceeded calculating all the possible combinations of t values using the in-built scipy function. We also used, as a proof, the alternative formula:

$$t= \frac{x_{1} - x_{2}}{\sqrt{\frac{\sigma ^2}{n_{1}}+\frac{\sigma ^2}{n_{1}}}}$$

We had to update the usual t student formula because our datasets(corresponding to all the possible tuples of boroughs) are **not** of the same length.

```python
import numpy as np
from scipy import stats
lista=['Bronx','Brooklyn','EWR','Manhattan','Queens','Staten Island','Unknown']
tstud=[]
plis=[]
a=JanBoroghData.groupby(['Borogh']).mean()
b=JanBoroghData.groupby(['Borogh']).std()
for m in range(0,6):
    for i in range(1,7):
        gg=lista[m]
        name=lista[i]
        c=JanBoroghData[JanBoroghData['Borogh']==gg]['Result']
        d=JanBoroghData[JanBoroghData['Borogh']==name]['Result']
        t2, p2 = stats.ttest_ind(c,d)
        tstud.append('tproof:'+str(t2))
        plis.append('pproof:'+str(p2))
    tstud.append('Next Borough')
    plis.append('Next Borogh')

```

The t test (also called Student’s T Test) compares two averages (means) and tells you if they are different from each other. The t test also tells you how significant the differences are. We now have to confront the p value with the t.student value and understand if we should reject the null hypothesis(Equal Mean of the two boroughs) or accept it.  

```python
#test between Bronx and Brookklyn on Result
print(str(tstud[0])+"   "+str(plis[0]))
#we get a  p value of 0.022 and thus we reject the null hypothesis and 
#thus it proves that the mean of the two distributions are different.
#We say that the p-value is statistically significant
```

```python
#test between Queens and Brooklyn on Result
print(str(tstud[3])+"   "+str(plis[3]))
#we get a good p value of 0.112 and thus we reject the null hypothesis and
#thus it proves that the mean of the two distributions are different.
#We say that the p-value is statistically significant
```

```python
#test between Staten Island and Brooklyn on Result
print(str(tstud[35])+"   "+str(plis[35]))
#we get a good p value of 0.0047 and thus we reject the null hypothesis and
#thus it proves that the mean of the two distributions are different.
#We say that the p-value is extremely significant
```
